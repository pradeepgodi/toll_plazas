```python
# Ensure CSV file exists with headers
def init_csv(file_path: str):
    if not os.path.exists(file_path):
        with open(file_path, mode='w', newline='', encoding='utf-8') as f:
            writer = csv.DictWriter(f, fieldnames=CSV_HEADERS)
            writer.writeheader()
```

```python
# Check for duplicates by name and lat/lng
existing_entries = set()
def is_duplicate(entry):
    # key = (entry['name'], entry['latitude'], entry['longitude'])
    key = (entry['latitude'], entry['longitude'])
    if key in existing_entries:
        return True
    existing_entries.add(key)
    return False
```

```python
# Save a record to CSV
def save_to_csv(record: Dict):
    if is_duplicate(record):
        return
    with open(OUTPUT_CSV, mode='a', newline='', encoding='utf-8') as f:
        writer = csv.DictWriter(f, fieldnames=CSV_HEADERS)
        writer.writerow(record)
```

```python
def extract_address(components):
    address = {"city": "N/A", "state": "N/A", "pincode": "N/A"}
    for comp in components:
        types = comp.get("types", [])
        if "locality" in types:
            address["city"] = comp.get("long_name", "N/A")
        if "administrative_area_level_1" in types:
            address["state"] = comp.get("long_name", "N/A")
        if "postal_code" in types:
            address["pincode"] = comp.get("long_name", "N/A")
    return address
```

```python
# Get phone number and other details via Place Details API (optional)
def get_place_details(place_id: str) -> Dict:
    details_url = f"https://maps.googleapis.com/maps/api/place/details/json?place_id={place_id}&\
                            fields=name,formatted_phone_number,geometry,address_components&key={API_KEY}"
    resp = requests.get(details_url).json()
    print("resp=",resp)
    phone = resp.get("result", {}).get("formatted_phone_number", "")
    extra_data=extract_address(resp['result']['address_components'])
    city=extra_data.get('city','')
    state=extra_data.get('state','')
    pincode=extra_data.get('pincode','')
    return {"phone": phone,"city":city,"state":state,"pincode":pincode}
```

```python
# Fetch weighbridges near given location
def fetch_weighbridges(lat: float, lng: float):
    next_page_token = None
    while True:
        params = {
            "location": f"{lat},{lng}",
            "radius": 50000,
            "keyword": "weighbridge",
            "key": API_KEY
        }
        if next_page_token:
            params['pagetoken'] = next_page_token

        response = requests.get(BASE_URL, params=params)
        
        data = response.json()
        # print(data['status'])
        if data['status']=='OK':

            for result in data.get("results", []):
                place_id = result.get("place_id")
                details = get_place_details(place_id)
                # print("result =",result)
                record = {
                    "name": result.get("name", ""),
                    "phone": details.get("phone", ""),
                    "city" :details.get("city", ""),
                    "state":details.get("state", ""),
                    "pincode":details.get("pincode", ""),
                    "business_status":result.get("business_status",""),
                    "latitude": result["geometry"]["location"]["lat"],
                    "longitude": result["geometry"]["location"]["lng"],
                    "formatted_address": result.get("vicinity", ""),
                    "place_id": result.get("place_id",""),
                    "rating":result.get("rating",""),
                    "user_ratings_total":result.get("user_ratings_total","")               
                }
                # print(record)
                save_to_csv(record)
    
            next_page_token = data.get("next_page_token")
            if not next_page_token:
                break
            time.sleep(2)
        else:
            raise Exception(data)
        
```

```python
# Generate a grid of lat/lng points for a given state boundary
def generate_grid(min_lat, max_lat, min_lng, max_lng, step=0.5):
    lat_lng_points = []
    lat = min_lat
    while lat <= max_lat:
        lng = min_lng
        while lng <= max_lng:
            lat_lng_points.append((lat, lng))
            lng += step
        lat += step
    return lat_lng_points
```

```python
# Searching at lat: 27.0, lng: 88.0
```

```python
# Main driver
def main():
    init_csv(OUTPUT_CSV)
    for state, bounds in STATE_BOUNDARIES.items():
        print(f"Fetching weighbridges for state: {state}")
        grid_points = generate_grid(
            bounds["min_lat"], bounds["max_lat"],
            bounds["min_lng"], bounds["max_lng"]
        )
        for lat, lng in grid_points:
            print(f"Searching at lat: {lat}, lng: {lng}")
            fetch_weighbridges(lat, lng)
            time.sleep(2)

if __name__ == "__main__":
    main()
```
